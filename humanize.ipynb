```python
def humanize_memory(memory):
    # transform memory into a human-readable string
    if isinstance(memory, Observation):
        return f"Before action {memory.timestamp + 1}, I saw: {memory.content}"
    elif isinstance(memory, Thought):
        return f"Before action {memory.timestamp + 1}, I thought: {memory.content}"
    elif isinstance(memory, Reflection):
        return f"Before action {memory.timestamp + 1}, I reflected: {memory.content}"
    # elif isinstance(memory, Plan):
    #     return f"Plan: {memory.content}"
    elif isinstance(memory, Action):
        return f"For action {memory.timestamp + 1}, I will: {memory.content}"
    else:
        return f"Memory: {memory}"
```

```python
for processing in range(0, 60):
    agent_files = sorted(glob(f"output/{processing}/agent_*.pkl"), key=lambda x: int(x.split("_")[1].replace(".pkl", "")))
    agent = pickle.load(open(agent_files[-1], "rb"))
    print(f"Processing {processing}:")
    print(f"  Memory length: {len(agent.memory.memories)}")
    with open(f"output/{processing}/humanized_memory.txt", "w") as f:
        f.write("\n".join([humanize_memory(m) for m in agent.memory.memories]))
```

```python
# processing = 7
for processing in range(0, 60):
    # humanize action
    # first read jsonl
    with open(f"output/{processing}/action_trace.txt", "r") as f:
        actions = [json.loads(line) for line in f.readlines()]
    result = []
    i = 0
    for action in actions:
        for a in action:
            result.append(f"Action {i+1}: {a['type']}, description: {a['description']}")
        i += 1
    with open(f"output/{processing}/humanized_action.txt", "w") as f:
        f.write("\n".join(result))
```
